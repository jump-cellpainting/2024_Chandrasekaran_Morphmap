```python
import pandas as pd
```

```python
orf_platemap_df = (
    pd.read_csv(
        "../00.download-and-process-annotations/input/experiment-metadata.tsv", sep="\t"
    )
    .query("Type=='ORF'")[["Plate_Map_Name", "Assay_Plate_Barcode"]]
    .rename(
        columns={
            "Plate_Map_Name": "Metadata_Platemap",
            "Assay_Plate_Barcode": "Metadata_Plate",
        }
    )
)
```

```python
well_df = pd.read_csv('../datasets/metadata/well.csv.gz')
plate_df = pd.read_csv('../datasets/metadata/plate.csv.gz')

well_plate_df = (
    well_df.merge(plate_df, how='left', on=['Metadata_Source','Metadata_Plate'])
)
```

```python
orf_well_plate_df = (
    (well_plate_df.query('Metadata_PlateType == "ORF"'))
    .reset_index(drop=True)
    .query("Metadata_Batch!='2021_08_23_Batch12'")
)

orf_well_plate_df
```

```python
orf_well_plate_platemap_df = (
    orf_well_plate_df.merge(orf_platemap_df, how='left', on=['Metadata_Plate'])
).drop_duplicates(subset=["Metadata_Platemap", "Metadata_Well"]).reset_index(drop=True)

orf_well_plate_platemap_df
```
