```python
import pandas as pd
```

```python
orf_df = pd.read_csv('output/orf.csv.gz', dtype=str)
orf_df["Metadata_NCBI_Gene_ID"] = orf_df["Metadata_NCBI_Gene_ID"].astype(str)

orf_df.head()
```

```python
crispr_df = pd.read_csv('output/crispr.csv.gz', dtype=str)
crispr_df["Metadata_NCBI_Gene_ID"] = crispr_df["Metadata_NCBI_Gene_ID"].astype(str)
crispr_df.head()
```

### Read metadata files

```python
corum_df = pd.read_csv("output/CORUM_proteincomplex_annotations.tsv", sep="\t").rename(
    columns={
        "Symbol": "Metadata_Approved_Symbol",
        "ComplexName": "Metadata_complexname",
    }
)
corum_df.head()
```

```python
class_df = pd.read_csv(
    "output/human_protein_atlas_class_annotations.tsv.gz", sep="\t"
).rename(
    columns={
        "Gene": "Metadata_Approved_Symbol",
        "protein_class": "Metadata_protein_class",
    }
)
class_df.head()
```

```python
disease_df = pd.read_csv(
    "output/human_protein_atlas_disease_annotations.tsv.gz", sep="\t"
).rename(
    columns={
        "Gene": "Metadata_Approved_Symbol",
        "disease_involvement": "Metadata_disease_involvement",
    }
)
disease_df.head()
```

```python
location_df = pd.read_csv(
    "output/human_protein_atlas_location_annotations.tsv.gz",
    sep="\t",
    usecols=["Gene", "subcellular_location"],
).rename(
    columns={
        "Gene": "Metadata_Approved_Symbol",
        "subcellular_location": "Metadata_subcellular_location",
    }
)
location_df.head()
```

```python
gene_family = pd.read_csv(
    "output/hgnc_approved_symbol_processed.tsv",
    sep="\t",
    usecols=["Approved_symbol", "Gene_group_ID"],
).rename(
    columns={
        "Approved_symbol": "Metadata_Approved_Symbol",
        "Gene_group_ID": "Metadata_gene_group_id",
    }
)
gene_family.head()
```

```python
homolog_df = pd.read_csv("output/homologues.tsv", sep="\t").rename(
    columns={"Symbol": "Metadata_Approved_Symbol", "Homologue": "Metadata_homologue"}
)

homolog_df.head()
```

```python
go_df = pd.read_csv(
    "output/gene-ontology-annotations.tsv.gz", sep="\t"
).rename(
    columns={
        "gene": "Metadata_Approved_Symbol",
        "go_bp_list": "Metadata_biological_process",
        "go_cc_list": "Metadata_cellular_component",
        "go_mf_list": "Metadata_molecular_function",
    }
)
go_df.head()
```

```python
wp_df = pd.read_csv("output/wikipathway-annotations.tsv.gz", sep="\t").rename(
    columns={
        "gene": "Metadata_Approved_Symbol",
        "wikipathway": "Metadata_wikipathway",
    }
)
wp_df.head()
```

```python
dependency_df = pd.read_csv(
    "output/CRISPRGeneDependency_U2OS.csv.gz", usecols=["geneID", "probability"]
).rename(
    columns={
        "geneID": "Metadata_NCBI_Gene_ID",
        "probability": "Metadata_dependency_probability",
    }
)

dependency_df.Metadata_NCBI_Gene_ID = dependency_df.Metadata_NCBI_Gene_ID.astype(str)
dependency_df.head()
```

```python
expression_df = pd.read_csv(
    "output/OmicsExpressionProteinCodingGenesTPMLogp1_U2OS.csv.gz",
    usecols=["geneID", "TPM"],
).rename(columns={"geneID": "Metadata_NCBI_Gene_ID", "TPM": "Metadata_TPM"})

expression_df.Metadata_NCBI_Gene_ID = expression_df.Metadata_NCBI_Gene_ID.astype(str)
expression_df.head()
```

```python
gene_effect_df = pd.read_csv(
    "output/CRISPRGeneEffect_U2OS.csv.gz", usecols=["geneID", "effect"]
).rename(columns={"geneID": "Metadata_NCBI_Gene_ID", "effect": "Metadata_gene_effect"})

gene_effect_df.Metadata_NCBI_Gene_ID = gene_effect_df.Metadata_NCBI_Gene_ID.astype(str)
gene_effect_df.head()
```

```python
orf_df = (
    orf_df.merge(corum_df, how="left", on="Metadata_Approved_Symbol")
    .merge(class_df, how="left", on="Metadata_Approved_Symbol")
    .merge(disease_df, how="left", on="Metadata_Approved_Symbol")
    .merge(location_df, how="left", on="Metadata_Approved_Symbol")
    .merge(gene_family, how="left", on="Metadata_Approved_Symbol")
    .merge(homolog_df, how="left", on="Metadata_Approved_Symbol")
    .merge(go_df, how="left", on="Metadata_Approved_Symbol")
    .merge(wp_df, how="left", on="Metadata_Approved_Symbol")
    .merge(dependency_df, how="left", on="Metadata_NCBI_Gene_ID")
    .merge(expression_df, how="left", on="Metadata_NCBI_Gene_ID")
    .merge(gene_effect_df, how='left', on="Metadata_NCBI_Gene_ID")
)

orf_df.count()
```

```python
crispr_df = (
    crispr_df.merge(corum_df, how="left", on="Metadata_Approved_Symbol")
    .merge(class_df, how="left", on="Metadata_Approved_Symbol")
    .merge(disease_df, how="left", on="Metadata_Approved_Symbol")
    .merge(location_df, how="left", on="Metadata_Approved_Symbol")
    .merge(gene_family, how="left", on="Metadata_Approved_Symbol")
    .merge(homolog_df, how="left", on="Metadata_Approved_Symbol")
    .merge(go_df, how="left", on="Metadata_Approved_Symbol")
    .merge(wp_df, how="left", on="Metadata_Approved_Symbol")
    .merge(dependency_df, how="left", on="Metadata_NCBI_Gene_ID")
    .merge(expression_df, how="left", on="Metadata_NCBI_Gene_ID")
    .merge(gene_effect_df, how='left', on="Metadata_NCBI_Gene_ID")
)

crispr_df.count()
```

```python
orf_df.to_csv('output/orf_metadata.tsv.gz', sep='\t', index=False)
crispr_df.to_csv('output/crispr_metadata.tsv.gz', sep='\t', index=False)
```
