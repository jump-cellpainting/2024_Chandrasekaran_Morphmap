```python
import pandas as pd
```

Download the essentiality file (CRISPRGeneDependency.csv) from the DepMap portal (https://depmap.org/portal/download/all/?releasename=DepMap+Public+23Q4) and save it outside the repository.

```python
dependency_df = pd.read_csv("../../CRISPRGeneDependency.csv").rename(
    columns={"Unnamed: 0": "ModelID"}
)
dependency_df.head()
```

This file contains many cell lines. I then filter the dataframe to just the row for U2OS. The `ModelID` for U2OS is "ACH-000364" (obtained from [CPJUMP0 design](https://github.com/jump-cellpainting/jump-cellpainting/blob/f778f3998dab15997607061a44ad438ac0f1f560/0.design-pilots/4.create-perturbation-list.Rmd#L355))

```python
dependency_df = (
    dependency_df.query('ModelID=="ACH-000364"')
    .transpose()
    .reset_index()
    .drop(index=0)
    .rename(columns={"index": "gene", 217: "probability"})
)
```

```python
dependency_df[["Symbol", "geneID"]] = dependency_df["gene"].str.split(" ", expand=True)
dependency_df = dependency_df.assign(geneID=lambda x: x["geneID"].str[1:-1])
```

```python
dependency_df[["geneID", "Symbol", "probability"]].to_csv(
    "output/CRISPRGeneDependency_U2OS.csv.gz", index=False
)
```
