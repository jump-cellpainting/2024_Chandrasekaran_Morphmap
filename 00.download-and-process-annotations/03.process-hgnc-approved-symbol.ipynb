```python
import pandas as pd
```

```python
hgnc_df = pd.read_csv("input/hgnc_approved_symbol.tsv", sep="\t").drop(
    columns=["Status"]
)
hgnc_df.head()
```

```python
hgnc_df = (
    hgnc_df.rename(
        columns={
            'Approved symbol': 'Approved_symbol',
            'Previous symbols': 'Previous_symbols',
            'Gene group ID': 'Gene_group_ID',
            'NCBI Gene ID': 'NCBI_Gene_ID'
        }
    )
)

hgnc_df["NCBI_Gene_ID"] = hgnc_df["NCBI_Gene_ID"].astype("Int64")
hgnc_df["NCBI_Gene_ID"] = hgnc_df["NCBI_Gene_ID"].astype("str")

hgnc_df.head()
```

```python
# Process Previous symbols

hgnc_df["Previous_symbols"] = hgnc_df["Previous_symbols"].str.split(", ").str.join('|')
hgnc_df.head()
```

```python
hgnc_df.to_csv('output/hgnc_approved_symbol_processed.tsv', sep='\t', index=False)
```
